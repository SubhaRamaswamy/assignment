```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def bernBits(n=100,outcomes=[-1,1]):
    s = []    
    for i in range(n):  
        a = np.random.choice(outcomes,p=[0.5,0.5])
        s.append(a)      
    return s
# bernBits()

def summation(toSum):
    return np.sum(toSum)

```

```python
N = [100,1000, 10000,10000]
for n in N:
    h = []
    k = []
    for j in xrange(n):
        h.append(bernBits())
        k.append(summation(toSum=bernBits()))
        
    tit = 'N='+str(n)+'  mean='+str(np.mean(k))+ '  variance='+str(np.var(k))
    plt.hist(k)
    plt.title(tit)
    plt.xlabel('sum')
    plt.ylabel('Frequency')
    plt.show()
    
```
