2(a):  Using Bernoulli bits, simulate a 10-bit random number generator that produces numbers uniformly distributed between 0 and 1. Plot the histogram of the output ofthis process and verify that it approaches a uniform distribution.

```python
def tenBit():
    s = []
    for i in range(10):  
        a = np.random.choice([0,1],p=[0.5,0.5])
        s.append(a)  
        joinStr= ''.join(map(str, s))
    return int(joinStr)

def  decify(strItem):
    return int(strItem,2)
```

```python
N = [100,1000,10000,100000]
for n in N:
    h = []
    for j in xrange(n):
        h.append(tenBit())
    # print h

    inAs = map(str,h)
    # print inAs

    decimal = map(decify,inAs)
    tit = 'N='+str(n)
    plt.hist(decimal)
    plt.title(tit)
    plt.xlabel('Numbers')
    plt.ylabel('Frequency')
    plt.show()
    
```
